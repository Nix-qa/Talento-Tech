```python
#aumentadno registro a la tabla

import pandas as pd

#create a dataset
#let´s build tha values

value_n_list = ["Andrew", "fhillip", "Jhon", "Raul", "danilo", "Alejandro", "Mario", "victor", "Juliana", "Milena"]
print(value_n_list)

value_w_list = [50, 78, 91.5, 100.5, 15, 33, 75, 45, 78, 80.4] #kg
value_h_list = [1.76,   1.87,   1.69,  1.70, 1.55, 1.90, 1.66, 1.70,  1.80, 1.85  ] #m

#values_bmi_list will be caculated: bmi = w/(h**2)
#value_bmi_list = 25,25,25, # init values kg/m

#let´s use a diccionary ({key:value})
#with the formt {"str:list"}
data_dict = {"name":value_n_list,
             "weight":value_w_list,
             "height":value_h_list} #,
             #"BMI":value_bmi_list}

print(data_dict)

#turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)
print(df)
#call a variable
df
```

agregando el indice corporal

```python
df["bmi"] = df["weight"]/(df["height"]**2)
df
```

agregando columna de clasificacion

```python
# Define a fuction to determine health condiction
def determine_health_condition(bmi):
    if 18.0 <= bmi <= 25.0:
        return "overweight"
    else:
        return "Normal"
    if 25.0 <= bmi <= 30.0:
        return "underweight"
    else:
        return "Normal"

# Apply the function to create the new colum
df["health_condition"] = df["bmi"].apply(determine_health_condition)
df
```

Creando Grafica de Coseno

```python
import math
import matplotlib.pyplot as plt

lista_eje_x = []
lista_eje_y = []

for i in range(0,720,1): # angulo en grados
    angulo=math.radians(i) # pasando el angulo a radianes
    x=angulo
    y=math.cos(x)
    lista_eje_x.append(x)
    lista_eje_y.append(y)

plt.plot(lista_eje_x,lista_eje_y)

plt.xlabel("Angulo (Radianes)")
plt.ylabel("Función Coseno: y=cos(x)")
plt.title("Grafica de la función de Coseno")
plt.show()
plt.savefig("gradica_coseno_x.jpg")
plt.close()
```

Creando grafica de funciones: lienal, cuadratica, cúbica, rais cuadrada

```python
import math as m
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace (0,2,100)
#generamos una grafica para una x lineal (recta)
plt.plot(x,x, label="Lineal") #y=x
#generamos una grafica para una x cuadratica (parabola)
plt.plot(x,x**2, label="Cuadratica") #y=x^2
#generamos una grafica para una x cubica
plt.plot(x,x**3, label= "cúbica") #y=x*x*x
#generamos una grafica para una x raiz cuadra (square root: sqrt)
plt.plot(x,x**(1/2), label="raiz cuadrada") #y=x**(1/2)
# agregamos las etiquetas y aádimos una leyenda
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graficas de funciones: lineal cuadratica, cúbica, raiz cuadrada (Creado por Danilo Enriquez)")
plt.legend()
plt.savefig("graficas_funciones_Danilo.png")
plt.show()
plt.close()
```

creando la misma grafica pero math.sqrt

```python

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace (-2,2,100)
#generamos una grafica para una x lineal (recta)
plt.plot(x,x, label="Lineal") #y=x
#generamos una grafica para una x cuadratica (parabola)
plt.plot(x,x**2, label="Cuadratica") #y=x^2
#generamos una grafica para una x cubica
plt.plot(x,x**3, label= "cúbica") #y=x*x*x
#generamos una grafica para una x raiz cuadra (square root: sqrt)
plt.plot(x,np.sqrt(x), label="raiz cuadrada") #y=x**(1/2)
# agregamos las etiquetas y añádimos una leyenda
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graficas de funciones: lineal cuadratica, cúbica, raiz cuadrada (Creado por Danilo Enriquez)")
plt.legend()
plt.savefig("graficas_funciones_Danilo.png")
plt.show()
plt.close()
```
